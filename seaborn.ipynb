```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x='total_bill', y='tip', data= tips)
```

```python
sns.histplot(tips['tip'], kde=  True, bins =15)
```

```python
# Create a histogram of the 'total_bill' column with hue based on 'smoker'
sns.histplot(data=tips, x='total_bill', hue='smoker', multiple='stack', bins=20)

```

# We add the hue parameter to differentiate bars based on the sex category.
This helps to compare the total_bill across different days, separately for males and females.

```python
# Create a bar plot with hue
sns.barplot(x='day', y='total_bill', hue='sex', data= tips)
```

```python
# Create a bar plot with custom confidence intervals
sns.barplot(x='day', y='total_bill', data=tips, ci=95)  # 95% confidence interval

```

```python
# Create a bar plot with median aggregation

import numpy as np
sns.barplot(x='day', y='total_bill', data=tips, estimator=np.median)
```

```python
# Create a box plot
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
#create a strip plot
sns.stripplot(x='day', y='total_bill', data= tips,hue = 'sex',dodge = True , jitter=True)
```

```python
# Create a joint plot with regression line
sns.jointplot(x='total_bill', y='tip', data=tips,kind='reg')
```

```python
sns.pairplot(tips)
```

```python
import pandas as pd
# Load the "tips" dataset
tips_data = sns.load_dataset("tips")

# Convert categorical variables to numerical using one-hot encoding
tips_data_encoded = pd.get_dummies(tips_data)

# Compute correlation matrix
corr_matrix = tips_data_encoded.corr()

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
# Load an example dataset
data = sns.load_dataset("iris")

# Create a cluster plot
sns.clustermap(data.drop(columns='species'), cmap='coolwarm', method='complete', standard_scale=1, row_cluster=True, col_cluster=True)
```
